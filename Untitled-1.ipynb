```python
# Define query parameters
parameters = {
    "results": 10,
    "gender": "male",
    "nat": "US"
}

# Make the API request
response = requests.get("https://randomuser.me/api/", params=parameters)

# Print the status code
print("Status Code:", response.status_code)

# Print the response content for debugging
print("Response Content:", response.text)

# If request is successful, process the response
if response.status_code == 200:
    try:
        # Convert the response to a Python dictionary
        data = json.loads(response.text)
        print("Type of data:", type(data))
        
        # Loop through the dictionary and print all first names
        for person in data['results']:
            print("First Name:", person['name']['first'])
        
        # Use pprint to print the data in a prettier format
        pprint(data)
    except json.JSONDecodeError:
        print("Error decoding JSON response")
else:
    print("Failed to retrieve data from randomuser.me API")
```

```python
# Make the API request to get people in space
astro_response = requests.get("http://api.open-notify.org/astros.json")

# Print the status code
print("Astronauts API Status Code:", astro_response.status_code)

# Print the response content for debugging
print("Astro Response Content:", astro_response.text)

# If request is successful, process the response
if astro_response.status_code == 200:
    try:
        # Convert the response to a Python dictionary
        astro_data = json.loads(astro_response.text)
        
        # Print the number of people in space
        print("Number of people in space:", astro_data['number'])
        
        # Loop through the dictionary and print all astronaut names
        for person in astro_data['people']:
            print("Astronaut Name:", person['name'])
        
        # Use pprint to print the data in a prettier format
        pprint(astro_data)
    except json.JSONDecodeError:
        print("Error decoding JSON response")
else:
    print("Failed to retrieve data from open-notify API")
```
